```python
def is2019(x):
    if str(x)[:4] == '2019':
        return True
    return False

amazon['is2019'] = amazon['Date'].apply(is2019)
```

```python
amazon.sample(10)
```

```python
apple['is2019'] = apple['Date'].apply(is2019)
facebook['is2019'] = facebook['Date'].apply(is2019)
google['is2019'] = google['Date'].apply(is2019)
netflix['is2019'] = netflix['Date'].apply(is2019)
```

```python
amazon = amazon[amazon['is2019']==True]
apple = apple[apple['is2019']==True]
facebook = facebook[facebook['is2019']==True]
google = google[google['is2019']==True]
netflix = netflix[netflix['is2019']==True]
```

```python
amazon['company'] = 'amazon'
apple['company'] = 'apple'
facebook['company'] = 'facebook'
google['company'] = 'google'
netflix['company'] = 'netflix'
```

```python
df = pd.concat([amazon,apple,facebook,google,netflix])
```

```python
df['Date']=df['Date'].apply(lambda x: str(x)[5:])
df['Date']=df['Date'].apply(lambda x: str(x[:2]+x[3:]))
df['Date']=df['Date'].apply(lambda x: int(x))
```
